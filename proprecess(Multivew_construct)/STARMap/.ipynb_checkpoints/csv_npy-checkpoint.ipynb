```python

csv_file_path = 'coord.csv'

# 输出提取坐标的文件名
output_coordinates_csv_file = 'extracted_coordinates.csv'
output_coordinates_npy_file = 'extracted_coordinates.npy' # 新增：NPY文件名
# --------------------------

try:
    # 1. 加载 CSV 数据
    # header=0 表示文件的第一行是表头。Pandas会自动使用这一行的值作为列名。
    print(f"尝试从 '{csv_file_path}' 加载数据...")
    df = pd.read_csv(csv_file_path, header=0)

    # 检查是否成功加载数据
    if df.empty:
        print(f"错误：文件 '{csv_file_path}' 为空或未能正确加载。")
    else:
        print("\n成功加载的原始数据 (前5行):")
        print(df.head())

        # 确保列名是 'X' 和 'Y' (或者你的CSV文件中的实际表头)
        # 如果你的CSV表头不是 'X' 和 'Y'，你需要在这里使用实际的列名
        expected_columns = ['X', 'Y']
        if not all(col in df.columns for col in expected_columns):
            print(f"\n错误：CSV文件中未找到预期的列名 {expected_columns}。实际列名: {df.columns.tolist()}")
            print("请确保CSV文件的表头与代码中引用的列名一致。")
        else:
            # 2. 提取坐标为 NumPy 数组
            # .to_numpy() 将DataFrame的两列转换为一个 NumPy 数组，形状为 (节点数, 2)
            # astype(float) 确保数据转换为浮点数类型，以防万一仍被识别为object
            coordinates = df[expected_columns].astype(float).to_numpy()

            print("\n提取的坐标数组 (前5行):")
            print(coordinates[:5])
            print(f"坐标数组的形状: {coordinates.shape}")
            print(f"坐标数组的数据类型: {coordinates.dtype}") # 检查数据类型

            # 确保坐标数据是数值类型
            if not np.issubdtype(coordinates.dtype, np.number):
                print("\n错误：坐标数据未能成功转换为数值类型。请再次检查CSV文件内容。")
            else:
                # --- 修改：保存提取的坐标矩阵为 CSV 和 NPY 格式 ---
                # 保存为 CSV
                try:
                    header_for_csv_output = 'X,Y'
                    np.savetxt(output_coordinates_csv_file, coordinates, delimiter=',', fmt='%.8f', header=header_for_csv_output, comments='')
                    print(f"\n提取的坐标数组已成功保存到 '{output_coordinates_csv_file}' (CSV格式)")
                except Exception as e_save_csv:
                    print(f"\n错误：保存提取的坐标数组到 '{output_coordinates_csv_file}' (CSV格式) 时发生错误: {e_save_csv}")

                # 保存为 NPY
                try:
                    np.save(output_coordinates_npy_file, coordinates)
                    print(f"提取的坐标数组已成功保存到 '{output_coordinates_npy_file}' (NPY格式)")
                except Exception as e_save_npy:
                    print(f"\n错误：保存提取的坐标数组到 '{output_coordinates_npy_file}' (NPY格式) 时发生错误: {e_save_npy}")
except FileNotFoundError:
    print(f"错误：文件 '{csv_file_path}' 未找到。请检查文件名和路径是否正确。")
except pd.errors.EmptyDataError:
    print(f"错误：文件 '{csv_file_path}' 是空的。")
except pd.errors.ParserError:
    print(f"错误：解析文件 '{csv_file_path}' 失败。请检查文件格式是否为有效的CSV (例如，确认分隔符是否为逗号)。")
except ValueError as ve:
    # astype(float) 或 cdist 可能会因为数据类型问题抛出 ValueError
    print(f"值错误: {ve}")
    print("这通常意味着坐标数据未能正确转换为数值类型，或者数组为空。请检查CSV内容和加载过程。")
except KeyError as ke:
    print(f"键错误: {ke}。这通常意味着CSV文件中缺少了 'X' 或 'Y' 列，或者表头名称不匹配。")
except Exception as e:
    print(f"发生了未知错误: {e}")
```
