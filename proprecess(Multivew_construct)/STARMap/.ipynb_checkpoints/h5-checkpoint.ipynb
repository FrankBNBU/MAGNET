```python
cell_names = adata_sdm.obs_names.tolist()

# 构建近距离的图 G_nearest_neighbors
G_full_nearest = nx.from_scipy_sparse_array(adata_sdm.obsp['nearest_neighbors'])

# 构建远距离的图 G_weight
G_weight = nx.from_scipy_sparse_array(adata_sdm.obsp['weight'])

# 创建映射字典：整数索引 -> 细胞名称
mapping_nearest = {i: name for i, name in enumerate(cell_names)}
mapping_weight = {i: name for i, name in enumerate(cell_names)}

# 重命名图的节点标签为细胞名称
G_full_nearest = nx.relabel_nodes(G_full_nearest, mapping_nearest)
G_weight = nx.relabel_nodes(G_weight, mapping_weight)

# 获取细胞类型信息
cell_types = adata.obs['cell_type']
cell_type_dict = cell_types.to_dict()  # 键为细胞名称，值为细胞类型

# 创建一个合并图 G_combined，只保留同类细胞之间的边
G_combined = nx.Graph()
```

```python
for u, v, data in G_full_nearest.edges(data=True):
    if cell_type_dict[u] == cell_type_dict[v]:  # 只保留同类细胞的边
        G_combined.add_edge(u, v, **data)

# 2. 然后处理远距离的边，如果该边在 G_combined 中不存在，才添加
for u, v, data in G_weight.edges(data=True):
    if cell_type_dict[u] == cell_type_dict[v]:  # 只保留同类细胞的边
        if not G_combined.has_edge(u, v):  # 仅在没有近距离边时添加远距离边
            G_combined.add_edge(u, v, **data)

# 调用 construct_boundary 函数
int_edges_new, int_edges_with_selfloop = spatial.construct_boundary(
    adata, 
    G=G_combined,
    weight_mat=None,  # 因为 G_combined 已经包含了权重信息
    domain_name='cell_type',  # 确保 domain_name 与您的细胞类型列名一致
    add_self_loops=True,  # 根据需要是否添加自环
    distance_aware=True   # 如果您希望考虑距离信息
)

# 查看生成的交互类型，确保只包含同一细胞类型的交互
print(int_edges_new['interaction'].unique())
```

```python
def prepare_data_list_intra_cell_type(
    count_df: pd.DataFrame,
    int_edges_new_with_selfloops: pd.DataFrame,
    lig_rec_pair_list,
    groups: list = None,
    summarization = "min",
    record_distance = True
) -> tuple:

    if groups is None:
        groups = int_edges_new_with_selfloops['interaction'].unique().tolist()
    
    data_list_dict = {}
    umi_sums = {}
    dist_list_dict = {}
    
    for g1 in tqdm.tqdm(groups):
        g1_dict = {}
        lr_pairs_g1 = int_edges_new_with_selfloops.loc[
            int_edges_new_with_selfloops['interaction'] == g1,
            ['cell1', 'cell2']
        ]
        dist_list_dict[g1] = int_edges_new_with_selfloops.loc[
            int_edges_new_with_selfloops['interaction'] == g1,
            'distance'
        ].values
        g11, g12 = g1.split('=')
        
        # 对于同一细胞类型内，g11 和 g12 是相同的
        # 计算 umi_sums
        _umi_sum = count_df.loc[lr_pairs_g1['cell1'].values, :].sum(axis=1).values
        g1_dict[g11] = _umi_sum.copy()
        g1_dict[g12] = _umi_sum.copy()  # g11 == g12
        
        data_list = []
        for lig, rec in lig_rec_pair_list:
            x = count_df.loc[lr_pairs_g1['cell1'].values, lig].values.astype('int')
            y = count_df.loc[lr_pairs_g1['cell2'].values, rec].values.astype('int')
            data_list.append((x, y))
        
        umi_sums[g1] = g1_dict.copy()
        data_list_dict[g1] = data_list.copy()
    
    if record_distance:
        return (data_list_dict, umi_sums, dist_list_dict)
    else:
        return (data_list_dict, umi_sums, None)
```

```python
# 获取所有细胞类型
cell_types = adata.obs['cell_type'].unique()

# 构建新的 int_edges_new_with_selfloops，仅包含同一细胞类型内的细胞对
int_edges_new_with_selfloops = pd.DataFrame(columns=['cell1', 'cell2', 'interaction', 'distance'])

for cell_type in cell_types:
    # 获取该细胞类型的细胞列表
    cells = adata.obs[adata.obs['cell_type'] == cell_type].index.tolist()
    
    # 构建所有可能的细胞对（可以根据需求调整，如仅考虑相邻细胞）
    cell_pairs = [(c1, c2) for c1 in cells for c2 in cells if c1 != c2]
    
    # 创建 DataFrame
    df = pd.DataFrame(cell_pairs, columns=['cell1', 'cell2'])
    df['interaction'] = f'{cell_type}={cell_type}'
    
    # 可选：计算细胞对之间的距离（如果有空间位置信息）
    # 这里假设 adata.obsm['spatial'] 存储了细胞的空间坐标
    if 'spatial' in adata.obsm.keys():
        spatial_coords = pd.DataFrame(adata.obsm['spatial'], index=adata.obs_names)
        df['distance'] = df.apply(lambda row: np.linalg.norm(
            spatial_coords.loc[row['cell1']] - spatial_coords.loc[row['cell2']]), axis=1)
    else:
        df['distance'] = 0  # 如果没有空间信息，设为 0 或其他默认值
    
    # 合并到总的 DataFrame 中
    int_edges_new_with_selfloops = pd.concat([int_edges_new_with_selfloops, df], ignore_index=True)
```

```python
def prepare_data_list_intra_cell_type(
    count_df: pd.DataFrame,
    int_edges_new_with_selfloops: pd.DataFrame,
    lig_rec_pair_list,
    groups: list = None,
    summarization = "min",
    record_distance = True
) -> tuple:

    if groups is None:
        groups = int_edges_new_with_selfloops['interaction'].unique().tolist()
    
    data_list_dict = {}
    umi_sums = {}
    dist_list_dict = {}
    
    for g1 in tqdm.tqdm(groups):
        g1_dict = {}
        lr_pairs_g1 = int_edges_new_with_selfloops.loc[
            int_edges_new_with_selfloops['interaction'] == g1,
            ['cell1', 'cell2']
        ]
        dist_list_dict[g1] = int_edges_new_with_selfloops.loc[
            int_edges_new_with_selfloops['interaction'] == g1,
            'distance'
        ].values
        g11, g12 = g1.split('=')
        
        # 对于同一细胞类型内，g11 和 g12 是相同的
        # 计算 umi_sums
        _umi_sum = count_df.loc[lr_pairs_g1['cell1'].values, :].sum(axis=1).values
        g1_dict[g11] = _umi_sum.copy()
        g1_dict[g12] = _umi_sum.copy()  # g11 == g12
        
        data_list = []
        for lig, rec in lig_rec_pair_list:
            x = count_df.loc[lr_pairs_g1['cell1'].values, lig].values.astype('int')
            y = count_df.loc[lr_pairs_g1['cell2'].values, rec].values.astype('int')
            data_list.append((x, y))
        
        umi_sums[g1] = g1_dict.copy()
        data_list_dict[g1] = data_list.copy()
    
    if record_distance:
        return (data_list_dict, umi_sums, dist_list_dict)
    else:
        return (data_list_dict, umi_sums, None)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import poisson, norm, multivariate_normal
from scipy.optimize import minimize
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import pairwise_distances
```

```python
distance_threshold = 50

# 计算细胞之间的距离矩阵
dist_matrix = pairwise_distances(coordinates)

# 构建邻接矩阵，连接距离在阈值内的细胞对
adjacency_matrix = (dist_matrix <= distance_threshold) & (dist_matrix > 0)

# 获取所有边 (i, j)
edges = np.argwhere(adjacency_matrix)

# 收集所有边上的配体和受体表达值对 (l_i, r_j)
l_i_values = ligand_expr[edges[:, 0]]
r_j_values = receptor_expr[edges[:, 1]]

# 计算每个细胞的总 UMI 计数，假设基因表达值为 UMI 计数
seqfish_data['Total_UMI'] = gene_expr.sum(axis=1)
total_umi = seqfish_data['Total_UMI'].values

# 获取对应的 N_i 和 N_j
N_i_values = total_umi[edges[:, 0]]
N_j_values = total_umi[edges[:, 1]]
```

```python
# 定义对数似然函数，用于估计参数
def negative_log_likelihood(params, l_i, r_j, N_i, N_j):
    mu_l, mu_r, rho = params
    # 确保 rho 在 (-1, 1) 内
    if not (-1 < rho < 1):
        return np.inf
    # 计算边缘分布的概率
    lambda_l = N_i * np.exp(mu_l)
    lambda_r = N_j * np.exp(mu_r)
    # 计算累积分布函数值
    u = poisson.cdf(l_i, lambda_l)
    v = poisson.cdf(r_j, lambda_r)
    # 防止 u 或 v 等于 0 或 1，导致无法计算对数
    u = np.clip(u, 1e-6, 1 - 1e-6)
    v = np.clip(v, 1e-6, 1 - 1e-6)
    # 转换为标准正态变量
    z_u = norm.ppf(u)
    z_v = norm.ppf(v)
    # 计算联合概率密度
    cov_matrix = [[1, rho], [rho, 1]]
    try:
        joint_pdf = multivariate_normal.logpdf(np.stack((z_u, z_v), axis=1), mean=[0, 0], cov=cov_matrix)
    except:
        return np.inf
    # 计算对数似然
    log_likelihood = np.sum(joint_pdf)
    return -log_likelihood

# 初始参数估计值
initial_params = [np.log(np.mean(l_i_values + 1)), np.log(np.mean(r_j_values + 1)), 0.0]

# 优化参数，最大化对数似然函数
result = minimize(negative_log_likelihood, initial_params, args=(l_i_values, r_j_values, N_i_values, N_j_values),
                  method='L-BFGS-B', bounds=[(None, None), (None, None), (-0.99, 0.99)])

# 获取估计的参数
mu_l_hat, mu_r_hat, rho_hat = result.x

print(f"Estimated parameters: mu_l = {mu_l_hat}, mu_r = {mu_r_hat}, rho = {rho_hat}")
```

```python
interaction_scores = []
for idx, (i, j) in enumerate(edges):
    l_i = l_i_values[idx]
    r_j = r_j_values[idx]
    N_i = N_i_values[idx]
    N_j = N_j_values[idx]
    # 边缘分布参数
    lambda_l = N_i * np.exp(mu_l_hat)
    lambda_r = N_j * np.exp(mu_r_hat)
    # 计算累积分布函数值
    u = poisson.cdf(l_i, lambda_l)
    v = poisson.cdf(r_j, lambda_r)
    # 防止 u 或 v 等于 0 或 1
    u = np.clip(u, 1e-6, 1 - 1e-6)
    v = np.clip(v, 1e-6, 1 - 1e-6)
    # 转换为标准正态变量
    z_u = norm.ppf(u)
    z_v = norm.ppf(v)
    z = np.array([z_u, z_v])
    # 协方差矩阵
    Sigma = np.array([[1, rho_hat], [rho_hat, 1]])
    # 计算 Mahalanobis 距离
    inv_Sigma = np.linalg.inv(Sigma)
    mahalanobis_distance = np.dot(z.T, np.dot(inv_Sigma, z))
    interaction_scores.append((i, j, mahalanobis_distance))

# 将结果转换为 DataFrame
interaction_df = pd.DataFrame(interaction_scores, columns=['Cell_Index_i', 'Cell_Index_j', 'Interaction_Score'])
```

```python
from joblib import Parallel, delayed
import numpy as np
from scipy.stats import rankdata, kendalltau, norm

def compute_pairwise_interaction(ligand_norm, receptor_norm, A, num_cells):
    """
    计算一个配体-受体对的细胞间相互作用评分
    """
    interaction_matrix = np.zeros((num_cells, num_cells))

    for i in range(num_cells):
        if ligand_norm[i] == 0:
            continue  # 跳过配体表达为 0 的细胞

        neighbors = np.where(A[i, :] > 0)[0]
        for j in neighbors:
            if receptor_norm[j] == 0:
                continue  # 跳过受体表达为 0 的细胞

            tau, _ = kendalltau([ligand_norm[i]], [receptor_norm[j]])
            if np.isnan(tau):
                tau = 0
            interaction_matrix[i, j] = tau

    return interaction_matrix

def compute_copulacci_interactions(X, A, lr_pair_indices, n_jobs=4):
    """
    计算细胞间相互作用评分矩阵。

    参数：
    - X: 基因表达矩阵，形状为 (细胞数, 基因数)
    - A: 邻接矩阵，形状为 (细胞数, 细胞数)
    - lr_pair_indices: 配体-受体对的基因索引列表
    - n_jobs: 并行处理的线程数量

    返回：
    - cci_matrix: 形状为 (细胞数, 细胞数, 配体-受体对数) 的相互作用评分矩阵
    """
    num_cells = X.shape[0]
    num_pairs = len(lr_pair_indices)
    cci_matrix = np.zeros((num_cells, num_cells, num_pairs))

    def process_lr_pair(lig_idx, rec_idx):
        ligand_expr = X[:, lig_idx]
        receptor_expr = X[:, rec_idx]

        # 估计边缘分布（使用排序转换为均匀分布）
        ligand_rank = rankdata(ligand_expr) / (len(ligand_expr) + 1)
        receptor_rank = rankdata(receptor_expr) / (len(receptor_expr) + 1)

        # 将均匀分布转换为标准正态分布（Gaussian Copula）
        ligand_norm = norm.ppf(ligand_rank + EPSILON)
        receptor_norm = norm.ppf(receptor_rank + EPSILON)

        return compute_pairwise_interaction(ligand_norm, receptor_norm, A, num_cells)

    # 并行计算每对配体-受体
    results = Parallel(n_jobs=n_jobs)(delayed(process_lr_pair)(lig_idx, rec_idx) for lig_idx, rec_idx in lr_pair_indices)

    # 将结果汇总到 cci_matrix 中
    for idx, result in enumerate(results):
        cci_matrix[:, :, idx] = result

    return cci_matrix
```

```python
def compute_copulacci_interactions(X, A, lr_pair_indices):
    """
    计算细胞间相互作用评分矩阵。

    参数：
    - X: 基因表达矩阵，形状为 (细胞数, 基因数)
    - A: 邻接矩阵，形状为 (细胞数, 细胞数)
    - lr_pair_indices: 配体-受体对的基因索引列表

    返回：
    - cci_matrix: 形状为 (细胞数, 细胞数, 配体-受体对数) 的相互作用评分矩阵
    """
    num_cells = X.shape[0]
    num_pairs = len(lr_pair_indices)
    cci_matrix = np.zeros((num_cells, num_cells, num_pairs))

    # 对每个配体-受体对，计算细胞间相互作用
    for idx, (lig_idx, rec_idx) in enumerate(lr_pair_indices):
        # 提取配体和受体的表达值
        ligand_expr = X[:, lig_idx]
        receptor_expr = X[:, rec_idx]

        # 估计边缘分布（使用排序转换为均匀分布）
        ligand_rank = rankdata(ligand_expr) / (len(ligand_expr) + 1)
        receptor_rank = rankdata(receptor_expr) / (len(receptor_expr) + 1)

        # 将均匀分布转换为标准正态分布（Gaussian Copula）
        ligand_norm = norm.ppf(ligand_rank + EPSILON)
        receptor_norm = norm.ppf(receptor_rank + EPSILON)

        # 计算细胞间的相互作用评分
        for i in range(num_cells):
            if ligand_expr[i] == 0:
                continue  # 跳过配体表达为 0 的细胞

            # 获取与细胞 i 相邻的细胞 j
            neighbors = np.where(A[i, :] > 0)[0]
            for j in neighbors:
                if receptor_expr[j] == 0:
                    continue  # 跳过受体表达为 0 的细胞

                # 计算 Copula 相互作用评分
                tau, _ = kendalltau(ligand_norm[i], receptor_norm[j])
                if np.isnan(tau):
                    tau = 0
                cci_matrix[i, j, idx] = tau

    return cci_matrix
```

```python
# 初始化有向图
G = nx.DiGraph()

# 添加节点（每个细胞）
cell_ids = adata.obs_names.tolist()
G.add_nodes_from(cell_ids)

# 计算每个细胞对的平均相互作用评分
num_cells = adata.n_obs
interaction_scores = cci_matrix.mean(axis=2)

# 设置相互作用评分的阈值，过滤弱相互作用
threshold = np.quantile(interaction_scores[interaction_scores > 0], 0.95)  # 取前5%的高评分

# 初始化列表存储结果
results = []

for i in range(num_cells):
    for j in range(num_cells):
        if A[i, j] == 1:
            mean_score = interaction_scores[i, j]
            if mean_score >= threshold:
                sender = cell_ids[i]
                receiver = cell_ids[j]
                G.add_edge(sender, receiver, weight=mean_score)
                results.append({
                    'sender_cell': sender,
                    'receiver_cell': receiver,
                    'interaction_score': mean_score
                })
9. 可视化单细胞相互作用的有向图
python
复制代码
# 将结果转换为 DataFrame
interaction_df = pd.DataFrame(results)

# 可视化
plt.figure(figsize=(12, 8))

# 使用 UMAP 坐标
if 'X_umap' not in adata.obsm:
    sc.pp.neighbors(adata)
    sc.tl.umap(adata)

pos = {cell_id: (adata.obsm['X_umap'][idx, 0], adata.obsm['X_umap'][idx, 1]) for idx, cell_id in enumerate(cell_ids)}

# 绘制节点
nx.draw_networkx_nodes(G, pos, node_size=10, node_color='lightblue')

# 绘制边
edges = G.edges()
weights = [G[u][v]['weight'] for u, v in edges]
nx.draw_networkx_edges(
    G, pos,
    edgelist=edges,
    width=[(w - threshold) * 10 for w in weights],
    arrowstyle='->',
    arrowsize=5,
    alpha=0.7
)

plt.title('单细胞相互作用有向图')
plt.axis('off')
plt.tight_layout()
plt.show()
完整代码汇总
python
复制代码
import numpy as np
import pandas as pd
from scipy.stats import rankdata, norm, kendalltau
from scipy.spatial.distance import pdist, squareform
import scanpy as sc
import squidpy as sq
import networkx as nx
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)

# 加载数据并预处理
adata = sq.datasets.seqfish()
sc.pp.log1p(adata)
sc.pp.filter_genes(adata, min_cells=3)

# 提取配体-受体对
import spatialdm as sdm
sdm.extract_lr(adata, 'mouse', min_cell=20)
lr_summary = adata.uns['lr_summary']
if lr_summary is None or lr_summary.empty:
    raise ValueError("未找到配体-受体对，请检查数据和参数设置。")
lr_pairs = lr_summary[['Ligand', 'Receptor']].drop_duplicates()

# 构建邻接矩阵
distances = squareform(pdist(adata.obsm['spatial']))
distance_threshold = 50
A = (distances <= distance_threshold).astype(int)
np.fill_diagonal(A, 0)

# 准备基因表达矩阵和基因索引
X = adata.X
if sp.issparse(X):
    X = X.toarray()

genes = adata.var_names.tolist()
genes_upper = [gene.upper() for gene in genes]
gene_to_idx = {gene: idx for idx, gene in enumerate(genes_upper)}

lr_pairs['ligand_upper'] = lr_pairs['Ligand'].str.upper()
lr_pairs['receptor_upper'] = lr_pairs['Receptor'].str.upper()
lr_pairs_filtered = lr_pairs[
    (lr_pairs['ligand_upper'].isin(gene_to_idx.keys())) &
    (lr_pairs['receptor_upper'].isin(gene_to_idx.keys()))
].copy()
lr_pairs_filtered['ligand_idx'] = lr_pairs_filtered['ligand_upper'].map(gene_to_idx)
lr_pairs_filtered['receptor_idx'] = lr_pairs_filtered['receptor_upper'].map(gene_to_idx)

# 实现 CopulaCCI 算法
def compute_copulacci_interactions(X, A, lr_pair_indices):
    num_cells = X.shape[0]
    num_pairs = len(lr_pair_indices)
    cci_matrix = np.zeros((num_cells, num_cells, num_pairs))
    EPSILON = 1e-10

    # 对每个配体-受体对，计算细胞间相互作用
    for idx, (lig_idx, rec_idx) in enumerate(lr_pair_indices):
        ligand_expr = X[:, lig_idx]
        receptor_expr = X[:, rec_idx]

        # 估计边缘分布
        ligand_rank = rankdata(ligand_expr) / (len(ligand_expr) + 1)
        receptor_rank = rankdata(receptor_expr) / (len(receptor_expr) + 1)

        ligand_norm = norm.ppf(ligand_rank + EPSILON)
        receptor_norm = norm.ppf(receptor_rank + EPSILON)

        for i in range(num_cells):
            if ligand_expr[i] == 0:
                continue
            neighbors = np.where(A[i, :] > 0)[0]
            for j in neighbors:
                if receptor_expr[j] == 0:
                    continue
                tau, _ = kendalltau([ligand_norm[i]], [receptor_norm[j]])
                if np.isnan(tau):
                    tau = 0
                cci_matrix[i, j, idx] = tau

    return cci_matrix

cci_matrix = compute_copulacci_interactions(X, A, lr_pairs_filtered[['ligand_idx', 'receptor_idx']].values.tolist())

# 构建单细胞相互作用的有向图
cell_ids = adata.obs_names.tolist()
G = nx.DiGraph()
G.add_nodes_from(cell_ids)
interaction_scores = cci_matrix.mean(axis=2)
threshold = np.quantile(interaction_scores[interaction_scores > 0], 0.95)

results = []
num_cells = adata.n_obs
for i in range(num_cells):
    for j in range(num_cells):
        if A[i, j] == 1:
            mean_score = interaction_scores[i, j]
            if mean_score >= threshold:
                sender = cell_ids[i]
                receiver = cell_ids[j]
                G.add_edge(sender, receiver, weight=mean_score)
                results.append({
                    'sender_cell': sender,
                    'receiver_cell': receiver,
                    'interaction_score': mean_score
                })

interaction_df = pd.DataFrame(results)

# 可视化
plt.figure(figsize=(12, 8))
if 'X_umap' not in adata.obsm:
    sc.pp.neighbors(adata)
    sc.tl.umap(adata)
pos = {cell_id: (adata.obsm['X_umap'][idx, 0], adata.obsm['X_umap'][idx, 1]) for idx, cell_id in enumerate(cell_ids)}
nx.draw_networkx_nodes(G, pos, node_size=10, node_color='lightblue')
edges = G.edges()
weights = [G[u][v]['weight'] for u, v in edges]
nx.draw_networkx_edges(
    G, pos,
    edgelist=edges,
    width=[(w - threshold) * 10 for w in weights],
    arrowstyle='->',
    arrowsize=5,
    alpha=0.7
)
plt.title('单细胞相互作用有向图')
plt.axis('off')
plt.tight_layout()
plt.show()
```
