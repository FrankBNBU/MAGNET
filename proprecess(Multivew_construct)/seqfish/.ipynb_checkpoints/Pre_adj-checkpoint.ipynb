```python
def visualize_celllevel_graph(df, title, edge_trace=None, edge_weights=1, publication=False):
    figsize_in_inches = (700 / 100, 650 / 100)  # 700x650 pixels at 100 DPI
    fig, ax = plt.subplots(figsize=figsize_in_inches)
    # 绘制细胞散点图，所有点的颜色统一为灰色
    ax.scatter(df["X"], df["Y"], color="gray", s=6, alpha=0.8, label="Cells")
    # 如果有边的坐标信息，添加边的描绘
    if edge_trace is not None:
        for i in range(0, len(edge_trace[0]), 2):
            x0, y0 = edge_trace[0][i], edge_trace[1][i]
            x1, y1 = edge_trace[0][i+1], edge_trace[1][i+1]
            ax.plot([x0, x1], [y0, y1], color='#888', lw=1.1, alpha=0.7)
    # 设置标题
    ax.set_title(title, fontsize=16, family="Times New Roman", color="black")
    ax.set_xlabel("X coordinate", fontsize=16, family="Times New Roman")
    ax.set_ylabel("Y coordinate", fontsize=16, family="Times New Roman")

    # 设置坐标轴样式
    ax.tick_params(axis='both', which='both', length=5, width=1, labelsize=12)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    # 设置出版风格的布局（清晰的白色背景，简洁的线条，去除不必要的元素）
    if not publication:
        ax.grid(False)  # 隐藏网格线
    # 如果需要出版风格，设置白色背景和没有图例
    else:
        ax.set_facecolor("white")
        ax.legend().set_visible(False)
    return fig
```

```python
def construct_celllevel_graph_with_distance(
    data_df,
    k, # 要连接的邻居数量 (用于构建 adjacency)
    m_neighbor_for_sigma=True, 
    default_global_sigma=70, 
    get_edges=False
):
    """
    进一步修改版：
    1. adjacency 矩阵 (N x N, 0/1) 基于每个细胞i的k个最强连接（强度S_ij基于sigma_i）。可能不对称。
    2. network_relationship (N x N) 记录所有原始计算的 i->j 权重 (S_ij, 基于 sigma_i)。此矩阵通常不对称。
    3. 可选的归一化步骤施加于最终的 network_relationship (如果选择对称归一化，会先对其进行对称化处理)。

    参数:
        data_df (pd.DataFrame): 包含细胞坐标 "X" 和 "Y" 列的DataFrame。
        k (int): 为每个细胞连接的邻居的数量（用于构建二进制 adjacency 矩阵）。
        m_neighbor_for_sigma (int, optional): ...
        default_global_sigma (float, optional): ...
        get_edges (bool, optional): ...
        normalize_method (str, optional): 对 network_relationship 的归一化方法。
            None (默认): 不进行归一化。
            "row_sum_to_1": 每行除以该行之和 (使每行和为1, 用于有向图随机游走)。
            "symmetric_normalized_laplacian_style": 先将 network_relationship 对称化 (A_sym = (A+A.T)/2), 
                                                  然后计算 D^(-1/2) * A_sym * D^(-1/2)。 (常用于无向图谱方法)
    返回:
        tuple: 包含:
            - adjacency (np.ndarray): num_cells x num_cells 数组 (0/1)。adjacency[i,j]=1 表示 j 是 i 的k个最强连接之一。
            - edges (list or None): 如果 get_edges 为 True，列表 [edge_x, edge_y] (对应 adjacency 中的边)。
            - edge_weights (list): 对应于 adjacency[i,j]=1 的每条有向边的原始高斯核强度 (S_ij)。
            - network_relationship (np.ndarray): num_cells x num_cells 数组。存储所有原始计算的 i->j 强度 S_ij (基于sigma_i)。
                                                 如果指定了归一化，则返回归一化后的版本。
    """
    num_cells = len(data_df)
    adjacency = np.zeros(shape=(num_cells, num_cells), dtype=int) 
    coords = np.vstack([data_df["X"].values, data_df["Y"].values]).T
    edge_x_list, edge_y_list = [], []
    calculated_strengths_for_selected_edges = [] 
    # network_relationship 将存储所有原始的 S_ij (基于 sigma_i)
    network_relationship = np.zeros((num_cells, num_cells))
    # --- 预计算每个细胞的局部sigma ---
    local_sigmas = np.full(num_cells, default_global_sigma)
    if m_neighbor_for_sigma is not None and m_neighbor_for_sigma > 0:
        m_rank_for_sigma = m_neighbor_for_sigma 
        for i_sig_calc in range(num_cells):
            current_cell_coords_for_sigma = coords[i_sig_calc]
            distances_for_sigma_calc = np.linalg.norm(coords - current_cell_coords_for_sigma, axis=1)
            sorted_original_indices = np.argsort(distances_for_sigma_calc)
            if len(sorted_original_indices) > m_rank_for_sigma: 
                mth_neighbor_original_idx = sorted_original_indices[m_rank_for_sigma]
                sigma_val = distances_for_sigma_calc[mth_neighbor_original_idx]
                if sigma_val > 1e-9: 
                    local_sigmas[i_sig_calc] = sigma_val
    # --- 步骤 B: 计算所有权重并填充 network_relationship, 然后筛选k个最强的连接以构建 adjacency ---
    for i in range(num_cells):
        x0, y0 = data_df["X"].values[i], data_df["Y"].values[i]
        current_cell_coords = coords[i]
        current_sigma_i = local_sigmas[i]
        euclidean_distances_from_i = np.linalg.norm(coords - current_cell_coords, axis=1)
        strengths_for_sorting = np.zeros(num_cells) 
        for j in range(num_cells):
            distance_i_to_j = euclidean_distances_from_i[j]
            strength_ij = 0.0
            if i == j:
                strength_ij = 1.0 # 自身与自身的原始强度通常定义为1 (或0)
                strengths_for_sorting[j] = -1.0 # 用于排序时排除自身被选为邻居
            else:
                if current_sigma_i > 1e-9: # 如果 sigma_i 有效
                    strength_ij = np.exp(- (distance_i_to_j ** 2) / (2 * current_sigma_i ** 2))
                elif distance_i_to_j < 1e-9: # 如果 sigma_i 接近于零 且 距离也接近于零 (例如重合点)
                    strength_ij = 1.0 
                # else (current_sigma_i 接近零，但距离不为零), strength_ij 保持为 0.0
                strengths_for_sorting[j] = strength_ij
            
            network_relationship[i, j] = strength_ij # 存储所有原始计算的 S_ij (基于 sigma_i)
        # 基于 strengths_for_sorting (其中 S_ii=-1) 来选择 k 个最强连接，用于构建 adjacency
        sorted_indices_by_strength = np.argsort(strengths_for_sorting)[::-1] 
        
        num_added_neighbors = 0
        for neighbor_idx in sorted_indices_by_strength:
            if num_added_neighbors == k: 
                break
            current_strength_for_neighbor = network_relationship[i, neighbor_idx] 

            if current_strength_for_neighbor > 1e-9: # 只选择强度大于阈值的连接
                adjacency[i, neighbor_idx] = 1 # 设置有向边 i -> neighbor_idx
                calculated_strengths_for_selected_edges.append(current_strength_for_neighbor)
                
                if get_edges: 
                    x1, y1 = data_df["X"].values[neighbor_idx], data_df["Y"].values[neighbor_idx]
                    edge_x_list.append(x0); edge_x_list.append(x1) 
                    edge_y_list.append(y0); edge_y_list.append(y1)
                num_added_neighbors += 1
        
    edges_for_return = [edge_x_list, edge_y_list] if get_edges else None

    return adjacency, edges_for_return, calculated_strengths_for_selected_edges, network_relationship
celllevel_adj1, edges1, edge_weights1, network_relationship1 = construct_celllevel_graph_with_distance(
    starting_df,
    k=5,
    m_neighbor_for_sigma=10,  
    default_global_sigma=70,
    get_edges=True
)

sim_fig = visualize_celllevel_graph(
    starting_df, title="Cell_Interaction(Spatial)", edge_trace=edges1, edge_weights=edge_weights1, publication=True
)
```

```python
def construct_celllevel_graph_with_LR(
    data_df, # 输入的数据框，应包含细胞的表达数据和空间坐标 (X, Y)
    ligand_receptor_pairs, # 一个包含多个 (配体列名, 受体列名) 元组的列表
    sigma=70, # 空间衰减因子中的 sigma 参数，控制距离对相互作用强度的影响范围
    k=2, # 为每个细胞选择 top-k 个最强的相互作用来构建邻接矩阵
    n_permutations=1000, # 置换检验的次数，用于评估观察到的相互作用得分的统计显著性
    alpha=0.05, # 统计显著性水平 (p值阈值)
    min_expression_threshold=0.01, # 判定细胞是否表达某个基因的最小表达量阈值
    get_edges=True # 是否返回用于绘图的边的坐标列表
):
    """
    构建基于统计显著性筛选的细胞级别相互作用图。
        - adjacency: (N x N) 0/1 矩阵，基于每个细胞的 top-k 相互作用。
        - network_relationship: (N x N) 全部显著 L-R 加权后的总得分矩阵。
        - edges_for_return: 如果 get_edges=True，则返回线条绘图坐标 [x_list, y_list]。
        - edge_weights: adjacency 中的有向边对应的得分列表。
    """
    num_cells = len(data_df) # 获取细胞的总数量
    coords = data_df[["X", "Y"]].values # 从数据框中提取所有细胞的X, Y坐标，并转换为numpy数组
    network_relationship = np.zeros((num_cells, num_cells)) # 初始化网络关系矩阵，大小为 (细胞数 x 细胞数)，所有元素初始为0                                                       
    # 遍历每一个提供的配体-受体对
    for ligand_col, receptor_col in ligand_receptor_pairs:
        L_vals = data_df[ligand_col].values # 获取数据框中指定配体列的所有细胞的表达值
        R_vals = data_df[receptor_col].values # 获取数据框中指定受体列的所有细胞的表达值

        # 创建布尔掩码，标记哪些细胞的配体/受体表达量达到了最小阈值
        expressed_ligand_mask = L_vals >= min_expression_threshold
        expressed_receptor_mask = R_vals >= min_expression_threshold

        # 初始化当前L-R对的相互作用得分矩阵
        current_lr_interaction_scores_for_this_pair = np.zeros((num_cells, num_cells))
        null_scores_population_list = [] # 用于存储通过置换检验产生的零假设下的得分，构建背景分布
        for _ in range(n_permutations):
            R_vals_permuted = np.random.permutation(R_vals) # 随机打乱当前受体的表达值顺序
            sample_size_for_null = num_cells * 1
            # 随机选择 sample_size_for_null 个细胞索引作为配体细胞 (允许重复选择)
            idx_i_for_null = np.random.choice(num_cells, sample_size_for_null, replace=True)
            # 随机选择 sample_size_for_null 个细胞索引作为受体细胞 (使用打乱后的受体表达值，允许重复选择)
            idx_j_for_null_perm = np.random.choice(num_cells, sample_size_for_null, replace=True)
            # 计算零假设下的相互作用得分 (配体细胞的真实表达值 * 打乱后的受体细胞的表达值)
            null_scores_population_list.extend(L_vals[idx_i_for_null] * R_vals_permuted[idx_j_for_null_perm])
        null_scores_population = np.array(null_scores_population_list) # 将零分布得分列表转换为numpy数组
        null_scores_population.sort() # 对零分布得分进行排序，方便后续计算p值时进行查找
        len_null_pop = len(null_scores_population) # 获取零分布的大小

        # --- 计算观察到的相互作用得分并评估其显著性 ---
        for i in range(num_cells): # 遍历所有细胞 i (作为潜在的发送细胞/配体细胞)
            if not expressed_ligand_mask[i]: # 如果细胞 i 不表达当前配体 (低于阈值)
                continue # 则跳过，不考虑其作为发送细胞
            for j in range(num_cells): # 遍历所有细胞 j (作为潜在的接收细胞/受体细胞)
                if i == j or not expressed_receptor_mask[j]: # 如果细胞 i 和 j 是同一个细胞，或者细胞 j 不表达当前受体
                    continue # 则跳过
                # 计算观察到的相互作用得分 (L_i * R_j)
                s_obs = L_vals[i] * R_vals[j]
                if s_obs < 1e-9: # 如果观察到的得分非常接近于0，则认为无相互作用
                    continue
                # 计算p值：(比s_obs更大或相等的零分布得分数量 + 1) / (零分布总大小 + 1)
                count_greater_equal = len_null_pop - np.searchsorted(null_scores_population, s_obs, side='left')
                p_value = (count_greater_equal + 1) / (len_null_pop + 1) # 加1是为了避免p值为0 (这是计算经验p值的一种方式)
                if p_value < alpha: # 如果计算出的p值小于预设的显著性水平alpha
                    # 该相互作用被认为是统计显著的
                    dist_sq = np.sum((coords[i] - coords[j]) ** 2) # 计算细胞 i 和细胞 j 之间空间距离的平方
                    # 应用空间衰减：距离越远，相互作用强度衰减越多。使用高斯衰减函数。
                    # sigma 参数控制衰减的程度，sigma越大，衰减越慢，相互作用的距离范围越大。
                    spatial_decay = np.exp(-dist_sq / (2 * sigma ** 2))
                    # 最终的相互作用强度 = 观察到的原始得分 * 空间衰减因子
                    current_lr_interaction_scores_for_this_pair[i, j] = s_obs * spatial_decay
        # 将当前L-R对计算得到的显著且经过空间衰减的相互作用得分，累加到总的网络关系矩阵中
        network_relationship += current_lr_interaction_scores_for_this_pair

    # --- 构建邻接矩阵 (Adjacency Matrix) ---
    # 邻接矩阵是一个0/1矩阵，表示细胞间是否存在基于top-k选择的强相互作用边
    adjacency = np.zeros((num_cells, num_cells), dtype=int) # 初始化邻接矩阵，数据类型为整数
    edge_x_list, edge_y_list, selected_edge_weights = [], [], [] # 初始化列表，用于存储边的坐标和权重

    for i in range(num_cells): # 再次遍历每个细胞 i，这次是确定其出边 (outgoing edges)
        strengths_from_i = network_relationship[i, :] # 获取细胞 i 指向其他所有细胞的累积相互作用强度

        # 找出细胞 i 的潜在相互作用邻居 (强度大于一个很小的值，且不是细胞自身)
        potential_neighbor_indices_with_strength = np.where(strengths_from_i > 1e-9)[0] # 获取强度大于阈值的邻居索引
        # 排除细胞自身作为邻居的可能性
        potential_neighbor_indices = potential_neighbor_indices_with_strength[potential_neighbor_indices_with_strength != i]

        if len(potential_neighbor_indices) == 0: # 如果细胞 i 没有潜在的相互作用邻居
            continue # 则跳过，处理下一个细胞

        actual_strengths = strengths_from_i[potential_neighbor_indices] # 获取这些潜在邻居的实际相互作用强度值
        # 确定要为细胞 i 选择的邻居数量：取 k 和实际潜在邻居数中的较小者
        num_neighbors_to_select = min(k, len(potential_neighbor_indices))

        if num_neighbors_to_select > 0: # 如果需要选择邻居
            # 从潜在邻居中，根据相互作用强度，选出最强的 num_neighbors_to_select 个邻居
            # np.argsort(actual_strengths) 返回的是排序后的索引（从小到大）
            # [-num_neighbors_to_select:] 取出最后num_neighbors_to_select个，即强度最大的那些邻居在potential_neighbor_indices中的局部索引
            top_k_local_indices_in_potential = np.argsort(actual_strengths)[-num_neighbors_to_select:]
            # 将这些局部索引转换为在全局细胞列表中的实际索引
            selected_neighbor_global_indices = potential_neighbor_indices[top_k_local_indices_in_potential]

            for neighbor_j_idx in selected_neighbor_global_indices: # 遍历选中的top-k邻居
                current_edge_weight = network_relationship[i, neighbor_j_idx] # 获取细胞 i 到邻居 j 的相互作用强度
                if current_edge_weight > 1e-9: # 再次确认强度有效
                    adjacency[i, neighbor_j_idx] = 1 # 在邻接矩阵中将 (i, j) 标记为1，表示存在一条从 i 到 j 的有向边
                    selected_edge_weights.append(current_edge_weight) # 将这条边的权重存起来
                    if get_edges: # 如果需要返回边的坐标 (用于后续绘图)
                        x0, y0 = data_df.iloc[i][["X", "Y"]] # 获取发送细胞 i 的X, Y坐标
                        x1, y1 = data_df.iloc[neighbor_j_idx][["X", "Y"]] # 获取接收细胞 j 的X, Y坐标
                        edge_x_list.extend([x0, x1]) # 将边的起点和终点x坐标加入列表
                        edge_y_list.extend([y0, y1]) # 将边的起点和终点y坐标加入列表
    # 准备返回结果
    edges_for_return = [edge_x_list, edge_y_list] if get_edges else None # 如果get_edges为True，则返回边坐标列表，否则返回None
    return adjacency,  edges_for_return, network_relationship,selected_edge_weights
```

```python
def construct_celllevel_graph_umap_inspired_standard_output(
    data_df, 
    k, 
    get_edges=True
):
    """
    基于 UMAP 原理构建细胞图，并输出为标准格式。

    :param data_df: pd.DataFrame : 包含 ["Cell_ID", "X", "Y"] 和特征列的 DataFrame
    :param k: int: 用于确定局部尺度 (sigma) 和最终邻接关系中每个细胞保留的最近邻个数
    :param get_edges: boolean: 是否返回用于可视化的边列表
    
    :return: tuple: 包含:
        - adjacency (np.ndarray): num_cells x num_cells 数组 (0/1)。adjacency[i,j]=1 表示 j 是 i 的k个
                                 最强连接之一 (基于 network_relationship 中的UMAP式综合评分 W_s)。
                                 此矩阵可能不对称 (因为是基于每行独立选k个最强)。
        - network_relationship (np.ndarray): num_cells x num_cells 数组。
                                             存储了对称化的UMAP式亲和力/相似度矩阵 W_s(i,j)。
                                             这个矩阵是构建 adjacency 的基础。
        - edges_for_return (list or None): 如果 get_edges 为 True，则为列表 [edge_x_list, edge_y_list]，
                                           包含用于绘制线条的坐标 (对应 adjacency[i,j]=1 的有向边)。
        - edge_weights (list): 对应于 adjacency[i,j]=1 的每条有向边的权重列表
                               (这些权重直接取自 network_relationship)。按处理顺序排列。
    """
    # --- 第1步：数据预处理 ---
    feature_df = data_df.drop(columns=["Cell_ID", "X", "Y", "Cell_Type"], errors='ignore')
    feature_df = feature_df.select_dtypes(include=[np.number]) 
    feature_matrix = feature_df.values
    num_cells = feature_matrix.shape[0]

    if num_cells == 0:
        print("数据行为空，无法继续。")
        empty_nxn_float = np.array([]).reshape(0,0).astype(float)
        empty_nxn_int = np.array([]).reshape(0,0).astype(int)
        empty_edges_list = [[], []] if get_edges else None
        return empty_nxn_int, empty_nxn_float, empty_edges_list, []

    # --- 第2步：数据标准化 (Z-Score) ---
    mean_vals = feature_matrix.mean(axis=0)
    std_devs = feature_matrix.std(axis=0)
    std_devs[std_devs == 0] = 1.0  
    feature_matrix_scaled = (feature_matrix - mean_vals) / std_devs
    feature_matrix_scaled = np.nan_to_num(feature_matrix_scaled) # 替换可能产生的NaN

    # --- 第3步：可选PCA降维 ---
    feature_matrix_processed = feature_matrix_scaled
    if num_cells > 1 :
        # n_components 必须 < n_samples 且 < n_features (对于某些情况)
        # PCA主成分数不能超过样本数或特征数中的较小者
        n_components_pca = min(50, feature_matrix_scaled.shape[1]) # 初步限制不超过50或原始特征数
        if num_cells <= n_components_pca : # 如果样本数不足以支持这么多主成分
             # n_components 必须严格小于 n_samples （对于某些sklearn版本或fit行为）
             # 至少需要1个样本才能有 num_cells-1 个主成分
            n_components_pca = max(1, num_cells -1) if num_cells > 1 else 1 


        if n_components_pca > 0 and feature_matrix_scaled.shape[1] > 0: #确保有特征和有效主成分数
            try:
                pca = PCA(n_components=n_components_pca, random_state=42)
                feature_matrix_processed = pca.fit_transform(feature_matrix_scaled)
            except ValueError as e:
                print(f"PCA执行出错 ({e})，将使用未降维的数据。可能是n_components ({n_components_pca}) 对于样本数 ({num_cells}) 或特征数 ({feature_matrix_scaled.shape[1]}) 不合适。")
                feature_matrix_processed = feature_matrix_scaled # 出错则回退
        else:
            # print(f"PCA跳过，因为计算出的有效主成分数 ({n_components_pca}) 或特征数不满足要求。")
            pass # feature_matrix_processed 保持为 scaled
            
    elif num_cells == 1:
        # print("仅有单个数据点，跳过PCA。")
        pass # feature_matrix_processed 保持为 scaled

    # --- 第4步：计算欧氏距离矩阵 ---
    if feature_matrix_processed.shape[0] > 1:
        dist_matrix = pairwise_distances(feature_matrix_processed, metric='euclidean')
    elif feature_matrix_processed.shape[0] == 1:
        dist_matrix = np.array([[0.0]])
    else: # 已被 num_cells == 0 捕获
        dist_matrix = np.array([]).reshape(0,0)


    # --- 第5步：确定自适应局部尺度 (sigmas 和 rhos) ---
    sigmas = np.zeros(num_cells)
    rhos = np.zeros(num_cells)

    if num_cells > 1:
        for i in range(num_cells):
            sorted_distances_from_i = np.sort(dist_matrix[i])
            rhos[i] = sorted_distances_from_i[1] # 到第1个最近邻的距离
            
            # sigma_i 是到第k个最近邻的距离
            # 确保 k 不超过存在的邻居数 (num_cells - 1)
            actual_k_for_sigma = min(k, num_cells - 1)
            if actual_k_for_sigma > 0 :
                sigma_val = sorted_distances_from_i[actual_k_for_sigma]
            else: # 如果 k=0 或只有一个细胞 (num_cells-1 = 0)
                # 如果还有其他细胞，用它们距离的平均值，否则用默认值
                sigma_val = np.mean(sorted_distances_from_i[1:]) if (num_cells - 1 > 0 and len(sorted_distances_from_i) > 1) else 1.0 
            sigmas[i] = max(sigma_val, 1e-10) # 加一个极小值避免sigma为0
    elif num_cells == 1: # 单个细胞的特殊处理
        sigmas[0] = 1.0 
        rhos[0] = 0.0   

    # --- 第6步：计算非对称连接概率 P(j|i) ---
    P_cond_val = np.zeros((num_cells, num_cells))
    if num_cells > 1:
        for i in range(num_cells):
            for j in range(num_cells):
                if i == j:
                    continue 
                dist_ij = dist_matrix[i, j]
                numerator_val = max(0, dist_ij - rhos[i])
                if sigmas[i] > 1e-12: # 避免除以非常小的 sigma_i
                    P_cond_val[i, j] = np.exp(-numerator_val / sigmas[i])

    # --- 第7步：对称化连接概率得到 network_relationship (原W_s) ---
    network_relationship = np.zeros((num_cells, num_cells)) # 这将是 W_s
    if num_cells > 1:
        for i in range(num_cells):
            # network_relationship 对角线可以保持为0，或根据需要设为1
            # UMAP 通常不强调对角线值，因为之后会排除自身进行k-NN选择
            # network_relationship[i, i] = 1.0 # 如果需要自身相似度为1 (可选)
            for j in range(i + 1, num_cells): 
                p_ji = P_cond_val[i, j]  # P(j | i)
                p_ij = P_cond_val[j, i]  # P(i | j)
                
                val = p_ji + p_ij - (p_ji * p_ij)
                network_relationship[i, j] = val
                network_relationship[j, i] = val 
    elif num_cells == 1: # 单细胞情况，network_relationship 是 [[0.]] 或 [[1.]]
        # network_relationship[0,0] = 1.0 # (可选)
        pass


    # --- 第8步：构建二进制邻接矩阵和绘图用的边 (符合新输出格式) ---
    adjacency = np.zeros((num_cells, num_cells), dtype=int)
    
    new_edge_x_list, new_edge_y_list = [], []
    new_edge_weights_list = []

    if num_cells > 0: # 确保有细胞可以处理
        for i in range(num_cells):
            if num_cells == 1 and k > 0: # 单细胞且要求邻居，无法选择
                break 

            neighbor_strengths = network_relationship[i].copy() # 从 W_s (即 network_relationship) 获取强度
            neighbor_strengths[i] = -np.inf # 排除自身，使其不被选为自己的邻居
            
            # 确定实际可选的邻居数量 (最多 num_cells - 1)
            num_potential_neighbors = num_cells - 1 if num_cells > 1 else 0
            actual_k_for_adj = min(k, num_potential_neighbors)
            
            if actual_k_for_adj > 0:
                # 获取权重最高的 actual_k_for_adj 个邻居的索引
                selected_neighbor_global_indices = np.argsort(neighbor_strengths)[-actual_k_for_adj:][::-1] # 降序
                
                for neighbor_idx in selected_neighbor_global_indices:
                    current_weight = network_relationship[i, neighbor_idx]
                    
                    if current_weight > 1e-9: # 仅当权重有意义时才建立连接
                        adjacency[i, neighbor_idx] = 1
                        new_edge_weights_list.append(current_weight)
                        
                        if get_edges:
                            # 确保 data_df["X"] 和 ["Y"] 存在且索引有效
                            x0, y0 = data_df["X"].values[i], data_df["Y"].values[i]
                            x1, y1 = data_df["X"].values[neighbor_idx], data_df["Y"].values[neighbor_idx]
                            new_edge_x_list.extend([x0, x1]) 
                            new_edge_y_list.extend([y0, y1])
    
    edges_for_return = [new_edge_x_list, new_edge_y_list] if get_edges else None
    if num_cells == 0 and get_edges: # 确保空数据时返回正确格式的空列表
        edges_for_return = [[], []]

    return adjacency, edges_for_return ,network_relationship , new_edge_weights_list
```

```python
def normalize_and_symmetrize(matrix):
    # 归一化到 0–1
    scaler = MinMaxScaler()
    matrix = scaler.fit_transform(matrix)
    # 对称化
    matrix = (matrix + matrix.T) / 2
    return matrix

# ========== 2. SNF 融合 ==========
def fuse_networks_with_snf(network_relationships_list, K_snf=20, t_snf=10):
    affinity_networks = [normalize_and_symmetrize(net) for net in network_relationships_list]
    try:
        fused_matrix = snf.snf(affinity_networks, K=K_snf, t=t_snf)
    except Exception as e:
        print(f"SNF failed with error: {e}, fallback to average.")
        fused_matrix = np.mean(affinity_networks, axis=0)
    return fused_matrix

# ========== 3. 构建邻接图 ==========
def build_graph_from_fused_matrix(data_df, fused_matrix, k=5, sim_threshold=0.0, get_edges=True):
    num_cells = len(data_df)
    combined_adjacency = np.zeros((num_cells, num_cells), dtype=int)
    combined_edge_x, combined_edge_y = [], []
    combined_edge_weights_list = []
    output_symmetric_weights_matrix = np.zeros((num_cells, num_cells), dtype=float)

    for i in range(num_cells):
        scores = fused_matrix[i, :].copy()
        scores[i] = -np.inf  # 排除自己
        neighbor_indices = np.argsort(scores)[::-1]

        count = 0
        for j in neighbor_indices:
            if count >= k:
                break
            if fused_matrix[i, j] >= sim_threshold:
                combined_adjacency[i, j] = 1
                if get_edges:
                    x0, y0 = data_df.loc[i, "X"], data_df.loc[i, "Y"]
                    x1, y1 = data_df.loc[j, "X"], data_df.loc[j, "Y"]
                    combined_edge_x.extend([x0, x1])
                    combined_edge_y.extend([y0, y1])
                    w = fused_matrix[i, j]
                    combined_edge_weights_list.append(w)
                    output_symmetric_weights_matrix[i, j] = w
                    output_symmetric_weights_matrix[j, i] = w
                count += 1

    edges_for_return = [combined_edge_x, combined_edge_y] if get_edges else None
    return combined_adjacency, edges_for_return, combined_edge_weights_list, output_symmetric_weights_matrix
```

```python
import numpy as np

# --- 函数定义 ---

def convert_adjacencylist2edgelist(adj_list_data, num_total_nodes=None):
    """
    将邻接表转换为边列表。
    适用于 adj_list_data 是一个 (N, k) 数组 (N个节点，每个节点k个邻居)
    或一个列表的列表。

    :param adj_list_data: 邻接表数据。
    :param num_total_nodes: 可选，图中总节点数，用于验证邻居索引。
    :return: nd.array, shape=(2, num_edges)
    """
    edge_list = []
    for node_idx, neighbors in enumerate(adj_list_data):
        for neighbor_idx in neighbors:
            is_valid_neighbor = isinstance(neighbor_idx, (int, np.integer)) and neighbor_idx >= 0
            if num_total_nodes is not None:
                is_valid_neighbor = is_valid_neighbor and neighbor_idx < num_total_nodes
            
            # 如果你的邻接表使用特定值（例如节点自身ID）表示无连接或填充，则需要额外过滤
            # 例如，如果不想添加自环，可以加上 node_idx != neighbor_idx
            # if is_valid_neighbor and node_idx != neighbor_idx: # 示例：过滤自环和无效索引

            if is_valid_neighbor: # 根据你的数据特性调整此条件
                edge_list.append([node_idx, int(neighbor_idx)]) #确保是int
            # 如果你的邻接表保证所有值都是有效的、想要的邻居索引，可以直接：
            # edge_list.append([node_idx, int(neighbor_idx)])

    if not edge_list:
        return np.array([]).reshape(2, 0) # 如果没有边，返回一个空的2x0数组
    return np.array(edge_list).T

def convert_adjacencymatrix2edgelist(adj_matrix_data):
    """
    将邻接矩阵转换为边列表。
    假设 adj_matrix_data[i, j] = 1 (或 True) 表示从 i 到 j 有一条边。

    :param adj_matrix_data: 邻接矩阵 (通常是布尔型或 0/1 整数型)。
    :return: nd.array, shape=(2, num_edges)
    """
    # np.where 返回两个数组：第一个是行索引 (源节点)，第二个是列索引 (目标节点)
    # 对于布尔矩阵，它会找到所有 True 的位置。
    # 对于整数矩阵，它会找到所有非零的位置（如果你的矩阵用1表示边）。
    source_nodes, target_nodes = np.where(adj_matrix_data)

    # 确保不包含自环，除非你的图允许自环
    # non_self_loops = source_nodes != target_nodes
    # source_nodes = source_nodes[non_self_loops]
    # target_nodes = target_nodes[non_self_loops]
    # 如果你的邻接矩阵已经是无自环的，则不需要上面这几行

    if len(source_nodes) == 0: # 如果没有边
        return np.array([]).reshape(2,0)
    return np.vstack((source_nodes, target_nodes)).astype(int)
def convert_adjacencylist2adjacencymatrix(adj_list):
    
    num_vertices = len(adj_list)
    adj_matrix = np.zeros(shape=(num_vertices,num_vertices))
     
    for i in range(num_vertices):
        for j in adj_list[i]:
            adj_matrix[i][j] = 1
            adj_matrix[j][i] = 1
     
    return adj_matrix
def convert_adjacencylist2adjacencymatrix(adj_list):
    
    num_vertices = len(adj_list)
    adj_matrix = np.zeros(shape=(num_vertices,num_vertices))
     
    for i in range(num_vertices):
        for j in adj_list[i]:
            adj_matrix[i][j] = 1
            adj_matrix[j][i] = 1
     
    return adj_matrix
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd # Assuming pandas is used for data_df

def construct_celllevel_graph_combined_corrected(
    data_df,
    network_relationships, # List of network_relationship matrices
    k,
    threshold=0.5,
    get_edges=True
):
    '''
    将多个网络关系矩阵合并为一个邻接图，考虑来自多个来源的交互。

    :param data_df: pd.DataFrame : 代表空间数据，包含 ["Cell_ID", "X", "Y"]
                                  Assumed to have a simple 0 to num_cells-1 integer index
                                  for .loc[i] to work as expected.
    :param network_relationships: list : 网络关系矩阵列表（如空间、配体-受体）
                                        Each matrix is num_cells x num_cells.
    :param k: int : 每个细胞的目标邻居数量上限
    :param threshold: float : 合并邻接矩阵的阈值（基于 aggregated_scores）
    :param get_edges: boolean : 是否返回边的坐标（用于可视化）

    :return: tuple:
        - combined_adjacency (np.ndarray): num_cells x num_cells 合并后的二进制 (0/1) 邻接矩阵。
                                           combined_adjacency[i,j]=1 表示 j 是 i 的一个选定邻居。
                                           此矩阵可能不对称。
        - edges_for_return (list or None): 如果 get_edges 为 True, 列表 [edge_x_list, edge_y_list]
        - combined_edge_weights (list): 对应于 combined_adjacency 中每条有向边的权重列表。
                                        权重是各 network_relationships 中对应边权重的总和。
        - combined_edge_weight_matrix (np.ndarray): num_cells x num_cells 合并后的边权重矩阵。
                                                     combined_edge_weight_matrix[i,j] 是 i 和 j 之间合并后的总权重。
                                                     此矩阵在代码中被明确设置为对称的。
    '''
    num_cells = len(data_df)

    combined_adjacency = np.zeros((num_cells, num_cells), dtype=int)
    combined_edge_x, combined_edge_y = [], []
    combined_edge_weights = []
    combined_edge_weight_matrix = np.zeros((num_cells, num_cells), dtype=float)

    for i in range(num_cells):

        aggregated_scores = np.sum([relationship[i, :] for relationship in network_relationships], axis=0)

        # 从候选邻居中排除细胞自身
        aggregated_scores[i] = -np.inf # Ensures cell i is not selected as its own neighbor here

        # 步骤2: 查找综合得分大于等于阈值的候选邻居的索引
        candidate_neighbor_indices = np.where(aggregated_scores >= threshold)[0]
        if len(candidate_neighbor_indices) > 0:
            # 获取这些候选邻居的实际得分
            scores_of_candidates = aggregated_scores[candidate_neighbor_indices]
            
            # 根据得分对候选邻居的 *内部索引* 进行降序排序
            #（即，得分最高的在前面）
            sorted_internal_indices_of_candidates = np.argsort(scores_of_candidates)[::-1] 
            
            # 按照排序后的顺序，从原始的 candidate_neighbor_indices 中选出最终的邻居
            # 并确保不超过 k 个
            num_to_select = min(k, len(sorted_internal_indices_of_candidates))
            final_selected_neighbor_indices = candidate_neighbor_indices[sorted_internal_indices_of_candidates[:num_to_select]]
        else:
            final_selected_neighbor_indices = np.array([]) #确保是数组以便后续使用
 
        #final_selected_neighbor_indices = candidate_neighbor_indices[:k]

        # 步骤3: 更新合并后的邻接矩阵
        if len(final_selected_neighbor_indices) > 0:
            combined_adjacency[i, final_selected_neighbor_indices] = 1

        # 步骤4: 收集边的细节用于可视化（可选）
        if get_edges:
            try:
                x0, y0 = data_df.loc[i, "X"], data_df.loc[i, "Y"]
            except KeyError: # Fallback if columns X, Y are missing or index issue
                print(f"Warning: Could not get X, Y for cell index {i}. Edge visualization might be affected.")
                x0, y0 = 0,0 # Default coordinates

            for ncell_idx in final_selected_neighbor_indices:
                try:
                    x1, y1 = data_df.loc[ncell_idx, "X"], data_df.loc[ncell_idx, "Y"]
                except KeyError:
                    print(f"Warning: Could not get X, Y for cell index {ncell_idx}. Edge visualization might be affected.")
                    x1, y1 = 0,0 # Default coordinates

                combined_edge_x.extend([x0, x1])
                combined_edge_y.extend([y0, y1])
                
                current_combined_strength = 0
                if network_relationships: # Check if the list is not empty
                    current_combined_strength = sum([relationship[i, ncell_idx] for relationship in network_relationships])
                
                combined_edge_weights.append(current_combined_strength)
                combined_edge_weight_matrix[i, ncell_idx] = current_combined_strength
                # 该合并权重矩阵被设置为对称的，即使邻接矩阵本身可能不是
                combined_edge_weight_matrix[ncell_idx, i] = current_combined_strength

    edges_for_return = [combined_edge_x, combined_edge_y] if get_edges else None
    if num_cells == 0 and get_edges:
        edges_for_return = [[], []]
        
    return combined_adjacency, edges_for_return, combined_edge_weights, combined_edge_weight_matrix
celllevel_adj5, edges5, combined_edge_weights, combined_edge_weight_matrix = construct_celllevel_graph_combined_corrected(
    starting_df, [network_relationship5], k=5, threshold=0.65, get_edges=True
)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

def generate_edges_combined_weights(adj_matrix, df_coords, sigma, weight_cutoff=0.0, combination_method="multiply"):
    """
    根据邻接矩阵和空间坐标生成边信息，支持多种权重组合方式。
    """
    final_edges_info = []
    num_cells = adj_matrix.shape[0]

    if num_cells != df_coords.shape[0]:
        raise ValueError("邻接矩阵的维度与坐标数据中的细胞数量不匹配。")

    coords_x = df_coords["X"].values
    coords_y = df_coords["Y"].values

    if combination_method == "add_then_normalize":
        potential_edges_raw_data = []
        all_raw_combined_weights = []

        for i in range(num_cells):
            for j in range(i + 1, num_cells):
                if adj_matrix[i, j] > 0:
                    distance = np.sqrt((coords_x[j] - coords_x[i])**2 + (coords_y[j] - coords_y[i])**2)
                    distance_gaussian_weight = np.exp(-(distance**2) / (2 * sigma**2))
                    raw_combined_weight = adj_matrix[i, j] + distance_gaussian_weight
                    potential_edges_raw_data.append({'idx_i': i, 'idx_j': j, 'raw_weight': raw_combined_weight})
                    all_raw_combined_weights.append(raw_combined_weight)

        if not all_raw_combined_weights:
            return []

        min_raw = np.min(all_raw_combined_weights)
        max_raw = np.max(all_raw_combined_weights)
        range_raw = max_raw - min_raw

        for edge_data in potential_edges_raw_data:
            raw_weight = edge_data['raw_weight']
            normalized_weight = 0.0
            if range_raw > 0:
                normalized_weight = (raw_weight - min_raw) / range_raw
            elif min_raw > 0:
                normalized_weight = 1.0

            if normalized_weight > weight_cutoff:
                final_edges_info.append({
                    'idx_start': edge_data['idx_i'],
                    'idx_end': edge_data['idx_j'],
                    'weight': normalized_weight
                })

    elif combination_method == "multiply":
        for i in range(num_cells):
            for j in range(i + 1, num_cells):
                if adj_matrix[i, j] > 0:
                    distance = np.sqrt((coords_x[j] - coords_x[i])**2 + (coords_y[j] - coords_y[i])**2)
                    distance_gaussian_weight = np.exp(-(distance**2) / (2 * sigma**2))
                    final_combined_weight = adj_matrix[i, j] * distance_gaussian_weight

                    if final_combined_weight > weight_cutoff:
                        final_edges_info.append({
                            'idx_start': i,
                            'idx_end': j,
                            'weight': final_combined_weight
                        })
    else:
        raise ValueError(f"未知的 combination_method: {combination_method}。请选择 'multiply' 或 'add_then_normalize'。")

    return final_edges_info


# 设置参数
SIGMA_VALUE = 200
CUTOFF_THRESHOLD = 0.9
CHOSEN_METHOD = "add_then_normalize"

output_dir = "output_edges_with_ids"
os.makedirs(output_dir, exist_ok=True)

file_prefix = f"edges_{CHOSEN_METHOD}_sigma{SIGMA_VALUE}_cutoff{str(CUTOFF_THRESHOLD).replace('.', 'p')}"
csv_filename = os.path.join(output_dir, f"{file_prefix}.csv")
npy_filename = os.path.join(output_dir, f"{file_prefix}.npy")

if 'network_relationship5' in locals() and 'starting_df' in locals():
    print(f"\n使用 '{CHOSEN_METHOD}' 方法结合权重...")

    if starting_df.shape[0] != network_relationship5.shape[0]:
        raise ValueError(f"starting_df 的细胞数 ({starting_df.shape[0]}) 与邻接矩阵 ({network_relationship6.shape[0]}) 不匹配。")

    generated_edges_info = generate_edges_combined_weights(
        network_relationship5,
        starting_df,
        sigma=SIGMA_VALUE,
        weight_cutoff=CUTOFF_THRESHOLD,
        combination_method=CHOSEN_METHOD
    )

    if generated_edges_info:
        print(f"生成了 {len(generated_edges_info)} 条边。")

        edges_data_for_saving = []
        plot_edge_x_coords = []
        plot_edge_y_coords = []
        plot_edge_weights = []

        max_cell_id_len = 30
        if 'Cell_ID' in starting_df.columns:
            try:
                max_cell_id_len = starting_df['Cell_ID'].astype(str).map(len).max() + 1
            except:
                print("警告: 无法动态确定 Cell_ID 最大长度，使用默认值。")

        for edge_info in generated_edges_info:
            idx_start = edge_info['idx_start']
            idx_end = edge_info['idx_end']
            weight = edge_info['weight']

            cell_id_start = starting_df['Cell_ID'].iloc[idx_start]
            x_start = starting_df['X'].iloc[idx_start]
            y_start = starting_df['Y'].iloc[idx_start]

            cell_id_end = starting_df['Cell_ID'].iloc[idx_end]
            x_end = starting_df['X'].iloc[idx_end]
            y_end = starting_df['Y'].iloc[idx_end]

            edges_data_for_saving.append({
                'Cell_ID_Start': cell_id_start, 'X_Start': x_start, 'Y_Start': y_start,
                'Cell_ID_End': cell_id_end, 'X_End': x_end, 'Y_End': y_end,
                'Weight': weight
            })

            plot_edge_x_coords.extend([x_start, x_end])
            plot_edge_y_coords.extend([y_start, y_end])
            plot_edge_weights.append(weight)

        # 保存 CSV
        try:
            pd.DataFrame(edges_data_for_saving).to_csv(csv_filename, index=False)
            print(f"边数据已保存到: {csv_filename}")
        except Exception as e:
            print(f"保存 CSV 时出错: {e}")

        # 保存 NPY
        try:
            dtype = [('Cell_ID_Start', f'U{max_cell_id_len}'), ('X_Start', 'f8'), ('Y_Start', 'f8'),
                     ('Cell_ID_End', f'U{max_cell_id_len}'), ('X_End', 'f8'), ('Y_End', 'f8'),
                     ('Weight', 'f8')]

            structured_array = np.array(
                [(e['Cell_ID_Start'], e['X_Start'], e['Y_Start'],
                  e['Cell_ID_End'], e['X_End'], e['Y_End'], e['Weight']) for e in edges_data_for_saving],
                dtype=dtype
            )
            np.save(npy_filename, structured_array)
            print(f"边数据已保存到: {npy_filename}")
        except Exception as e:
            print(f"保存 NPY 时出错: {e}")

        # 可视化
        title_suffix = f"Method: {CHOSEN_METHOD}, Sigma: {SIGMA_VALUE}, Cutoff: {CUTOFF_THRESHOLD}"
        combined_edges_trace_for_plot = [plot_edge_x_coords, plot_edge_y_coords]
        try:
            final_fig = visualize_celllevel_graph(
                starting_df,
                title=f"Cell Interaction ({title_suffix})",
                edge_trace=combined_edges_trace_for_plot,
                edge_weights=plot_edge_weights,
                publication=True
            )
            plt.show()
            if plot_edge_weights:
                print(f"保留的边权重范围: min={min(plot_edge_weights):.4f}, max={max(plot_edge_weights):.4f}")
        except NameError as e:
            if 'visualize_celllevel_graph' in str(e):
                print("错误: `visualize_celllevel_graph` 函数未定义。")
            else:
                raise e
        except Exception as e:
            print(f"可视化时出错: {e}")
    else:
        print(f"在 sigma={SIGMA_VALUE} 和 cutoff={CUTOFF_THRESHOLD} 下，没有生成有效边。")
else:
    print("错误: 'network_relationship6' 或 'starting_df' 未定义。")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

def generate_edges_combined_weights(adj_matrix, df_coords, sigma, weight_cutoff=0.0, combination_method="multiply"):
    """
    根据邻接矩阵和空间坐标生成边信息，支持多种权重组合方式。
    """
    final_edges_info = []
    num_cells = adj_matrix.shape[0]

    if num_cells != df_coords.shape[0]:
        raise ValueError("邻接矩阵的维度与坐标数据中的细胞数量不匹配。")

    coords_x = df_coords["X"].values
    coords_y = df_coords["Y"].values

    if combination_method == "add_then_normalize":
        potential_edges_raw_data = []
        all_raw_combined_weights = []

        for i in range(num_cells):
            for j in range(i + 1, num_cells):
                if adj_matrix[i, j] > 0:
                    distance = np.sqrt((coords_x[j] - coords_x[i])**2 + (coords_y[j] - coords_y[i])**2)
                    distance_gaussian_weight = np.exp(-(distance**2) / (2 * sigma**2))
                    raw_combined_weight = adj_matrix[i, j] + distance_gaussian_weight
                    potential_edges_raw_data.append({'idx_i': i, 'idx_j': j, 'raw_weight': raw_combined_weight})
                    all_raw_combined_weights.append(raw_combined_weight)

        if not all_raw_combined_weights:
            return []

        min_raw = np.min(all_raw_combined_weights)
        max_raw = np.max(all_raw_combined_weights)
        range_raw = max_raw - min_raw

        for edge_data in potential_edges_raw_data:
            raw_weight = edge_data['raw_weight']
            normalized_weight = 0.0
            if range_raw > 0:
                normalized_weight = (raw_weight - min_raw) / range_raw
            elif min_raw > 0:
                normalized_weight = 1.0

            if normalized_weight > weight_cutoff:
                final_edges_info.append({
                    'idx_start': edge_data['idx_i'],
                    'idx_end': edge_data['idx_j'],
                    'weight': normalized_weight
                })

    elif combination_method == "multiply":
        for i in range(num_cells):
            for j in range(i + 1, num_cells):
                if adj_matrix[i, j] > 0:
                    distance = np.sqrt((coords_x[j] - coords_x[i])**2 + (coords_y[j] - coords_y[i])**2)
                    distance_gaussian_weight = np.exp(-(distance**2) / (2 * sigma**2))
                    final_combined_weight = adj_matrix[i, j] * distance_gaussian_weight

                    if final_combined_weight > weight_cutoff:
                        final_edges_info.append({
                            'idx_start': i,
                            'idx_end': j,
                            'weight': final_combined_weight
                        })
    else:
        raise ValueError(f"未知的 combination_method: {combination_method}。请选择 'multiply' 或 'add_then_normalize'。")

    return final_edges_info


# 设置参数
SIGMA_VALUE = 100
CUTOFF_THRESHOLD = 0.5
CHOSEN_METHOD = "multiply"

output_dir = "output_edges_with_ids"
os.makedirs(output_dir, exist_ok=True)

file_prefix = f"edges_{CHOSEN_METHOD}_sigma{SIGMA_VALUE}_cutoff{str(CUTOFF_THRESHOLD).replace('.', 'p')}"
csv_filename = os.path.join(output_dir, f"{file_prefix}.csv")
npy_filename = os.path.join(output_dir, f"{file_prefix}.npy")

if 'network_relationship6' in locals() and 'starting_df' in locals():
    print(f"\n使用 '{CHOSEN_METHOD}' 方法结合权重...")

    if starting_df.shape[0] != network_relationship6.shape[0]:
        raise ValueError(f"starting_df 的细胞数 ({starting_df.shape[0]}) 与邻接矩阵 ({network_relationship6.shape[0]}) 不匹配。")

    generated_edges_info = generate_edges_combined_weights(
        network_relationship6,
        starting_df,
        sigma=SIGMA_VALUE,
        weight_cutoff=CUTOFF_THRESHOLD,
        combination_method=CHOSEN_METHOD
    )

    if generated_edges_info:
        print(f"生成了 {len(generated_edges_info)} 条边。")

        edges_data_for_saving = []
        plot_edge_x_coords = []
        plot_edge_y_coords = []
        plot_edge_weights = []

        max_cell_id_len = 30
        if 'Cell_ID' in starting_df.columns:
            try:
                max_cell_id_len = starting_df['Cell_ID'].astype(str).map(len).max() + 1
            except:
                print("警告: 无法动态确定 Cell_ID 最大长度，使用默认值。")

        for edge_info in generated_edges_info:
            idx_start = edge_info['idx_start']
            idx_end = edge_info['idx_end']
            weight = edge_info['weight']

            cell_id_start = starting_df['Cell_ID'].iloc[idx_start]
            x_start = starting_df['X'].iloc[idx_start]
            y_start = starting_df['Y'].iloc[idx_start]

            cell_id_end = starting_df['Cell_ID'].iloc[idx_end]
            x_end = starting_df['X'].iloc[idx_end]
            y_end = starting_df['Y'].iloc[idx_end]

            edges_data_for_saving.append({
                'Cell_ID_Start': cell_id_start, 'X_Start': x_start, 'Y_Start': y_start,
                'Cell_ID_End': cell_id_end, 'X_End': x_end, 'Y_End': y_end,
                'Weight': weight
            })

            plot_edge_x_coords.extend([x_start, x_end])
            plot_edge_y_coords.extend([y_start, y_end])
            plot_edge_weights.append(weight)

        # 保存 CSV
        try:
            pd.DataFrame(edges_data_for_saving).to_csv(csv_filename, index=False)
            print(f"边数据已保存到: {csv_filename}")
        except Exception as e:
            print(f"保存 CSV 时出错: {e}")

        # 保存 NPY
        try:
            dtype = [('Cell_ID_Start', f'U{max_cell_id_len}'), ('X_Start', 'f8'), ('Y_Start', 'f8'),
                     ('Cell_ID_End', f'U{max_cell_id_len}'), ('X_End', 'f8'), ('Y_End', 'f8'),
                     ('Weight', 'f8')]

            structured_array = np.array(
                [(e['Cell_ID_Start'], e['X_Start'], e['Y_Start'],
                  e['Cell_ID_End'], e['X_End'], e['Y_End'], e['Weight']) for e in edges_data_for_saving],
                dtype=dtype
            )
            np.save(npy_filename, structured_array)
            print(f"边数据已保存到: {npy_filename}")
        except Exception as e:
            print(f"保存 NPY 时出错: {e}")

        # 可视化
        title_suffix = f"Method: {CHOSEN_METHOD}, Sigma: {SIGMA_VALUE}, Cutoff: {CUTOFF_THRESHOLD}"
        combined_edges_trace_for_plot = [plot_edge_x_coords, plot_edge_y_coords]
        try:
            final_fig = visualize_celllevel_graph(
                starting_df,
                title=f"Cell Interaction ({title_suffix})",
                edge_trace=combined_edges_trace_for_plot,
                edge_weights=plot_edge_weights,
                publication=True
            )
            plt.show()
            if plot_edge_weights:
                print(f"保留的边权重范围: min={min(plot_edge_weights):.4f}, max={max(plot_edge_weights):.4f}")
        except NameError as e:
            if 'visualize_celllevel_graph' in str(e):
                print("错误: `visualize_celllevel_graph` 函数未定义。")
            else:
                raise e
        except Exception as e:
            print(f"可视化时出错: {e}")
    else:
        print(f"在 sigma={SIGMA_VALUE} 和 cutoff={CUTOFF_THRESHOLD} 下，没有生成有效边。")
else:
    print("错误: 'network_relationship6' 或 'starting_df' 未定义。")
```
